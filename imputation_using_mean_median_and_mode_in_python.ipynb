```python
import pandas as pd
df=pd.read_table("p015036_sepsis_sofa_qsofa_with_nan.psv",sep="|") 
```

```python
df.head()
```

```python
df1=df[['Resp','SBP','Platelets','Bilirubin_total','Creatinine','ICULOS','SepsisLabel']]
```

```python
df1.head()
```

# Imputation using mean

```python
df1.mean()
```

```python
import numpy as np
df_mean=df1.replace(np.nan, df1.mean())
```

```python
# Running sofa and qsofa 
result_qsofa = []
result_sofa= []
        
for a,b,c,d,e in zip(df_mean["Resp"], df_mean["SBP"], df_mean["Platelets"], df_mean["Bilirubin_total"], df_mean["Creatinine"]): 
    if a>=22 or b<=100: 
        result_qsofa.append(1)
        if c<100 or d>=2 or e>=2: 
            result_sofa.append(1)
        else:
            result_sofa.append(0)
    else:
        result_qsofa.append(0) 
        result_sofa.append(0)
        
df_mean['qSOFA']=result_qsofa
df_mean['SOFA']=result_sofa

df_mean
```

# Imputation using median 

```python
df1.median()
```

```python
import numpy as np
df_median=df1.replace(np.nan, df1.median())
```

```python
# Running sofa and qsofa 
result_qsofa = []
result_sofa= []
        
for a,b,c,d,e in zip(df_median["Resp"], df_median["SBP"], df_median["Platelets"], df_median["Bilirubin_total"], df_median["Creatinine"]): 
    if a>=22 or b<=100: 
        result_qsofa.append(1)
        if c<100 or d>=2 or e>=2: 
            result_sofa.append(1)
        else:
            result_sofa.append(0)
    else:
        result_qsofa.append(0) 
        result_sofa.append(0)
        
df_median['qSOFA']=result_qsofa
df_median['SOFA']=result_sofa

df_median
```

# Imputation using mode 

```python
df1.mode().iloc[0]
```

```python
df_mode=df1.fillna(df1.mode().iloc[0])
```

```python
# Running sofa and qsofa 
result_qsofa = []
result_sofa= []
        
for a,b,c,d,e in zip(df_mode["Resp"], df_mode["SBP"], df_mode["Platelets"], df_mode["Bilirubin_total"], df_mode["Creatinine"]): 
    if a>=22 or b<=100: 
        result_qsofa.append(1)
        if c<100 or d>=2 or e>=2: 
            result_sofa.append(1)
        else:
            result_sofa.append(0)
    else:
        result_qsofa.append(0) 
        result_sofa.append(0)
        
df_mode['qSOFA']=result_qsofa
df_mode['SOFA']=result_sofa

df_mode
```

```python
i=df[['SepsisLabel','ICULOS','qSOFA','SOFA']]
j=df_mean[['SepsisLabel','ICULOS','qSOFA','SOFA']]
k=df_median[['SepsisLabel','ICULOS','qSOFA','SOFA']]
l=df_mode[['SepsisLabel','ICULOS','qSOFA','SOFA']] 
```

```python
df_all = pd.concat([i.set_index('ICULOS'),j.set_index('ICULOS'),k.set_index('ICULOS'), l.set_index('ICULOS') ], 
                   axis='columns', keys=['with_nan', 'with_mean', "with_median", "with_mode"])
df_all
```
