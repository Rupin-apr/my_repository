```python
import os
import pandas as pd
directory='/home/ibab/Pictures/working_on_small_subset_for_sofa_qsofa_scoring'

for filename in os.listdir(directory):
    if filename.endswith(".psv"):
        data=pd.read_table(filename, sep='|', engine='python')
        result_qsofa = []
        result_sofa= []
        
        for a,b,c,d,e in zip(data["Resp"], data["SBP"],data["Platelets"],data["Bilirubin_total"],data["Creatinine"]): 
            if a>=22 or b<=100: 
                result_qsofa.append(1)
                if c<100 or d>=2 or e>=2: 
                    result_sofa.append(1)
                else:
                    result_sofa.append(0)
            else:
                result_qsofa.append(0)
                result_sofa.append(0)
        
        data['qSOFA']=result_qsofa
        data['SOFA']=result_sofa
        print(data)
        data.to_csv(filename,sep='|')

```
