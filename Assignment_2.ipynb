```
import pandas as pd
df=pd.read_csv("diabetes.csv")
df.head()
```

```
import matplotlib.pyplot as plt
# univariate plot eg-histogram
# bivariate plot eg-scatter

df.boxplot(by='Age', column=["BloodPressure"], grid =False, figsize=(20,10))



```

```
# To see how Glucose & insulin are related:
# Making bivariate plot:
import matplotlib.pyplot as plt
df.plot.bar(x="Glucose",y="Insulin")
```

```
x= df.groupby('BMI')['Glucose'].mean().sort_values().tail(15)

x.plot(kind='barh')
```

```
# scatter plot:
df.plot.scatter( x='BloodPressure',y='Insulin')
```

```
# scatter plot:
df.plot.scatter(x='Pregnancies',y='BMI',c='g',marker=r'$\clubsuit$')




```
